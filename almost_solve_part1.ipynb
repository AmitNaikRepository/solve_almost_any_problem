```
#now lets build the most simple algorithm with one hot encoding 
import pandas as pd 
import numpy as np 

from sklearn import linear_model 
from sklearn import metrics 
from sklearn import preprocessing


def run(fold):
  #load the same training data with folds 
  df=pd.read_csv('the_kfold.csv')
  
  #all column jsut ignor the id , targets, kfold 

  feature =[f for f in df.columns if f not in ('id','target','kfold')]



  #fill the nan of the system with None 
  for col in features:
    df.loc[:,col]=df[col].astype(str).fillna('NONE')

    #now get hte training data on the fold 
    df_train=df[df.kfold!=fold].reset_index(drop=True)

    #now get the data using the validation 
    df_valid=df[df.kfold!=fold].reset_index(drop=True)

    #initialize the one hot encoding to the system 
    ohe=preprocessing.OneHotEncoding()

    #now fit the training data and testing data 

    full_data=pd.concat(df_train[features],df_valid[features])

    #Now fit the Model 
    ohe.fit(full_data[features])

    #transform the training data 

    x_train=ohe.transform(df_train[features])

    x_valid=ohe.transform(df_valid[feature])


    model=linear_model.Logistic_regression()

    model.fit(x_train,df_train.target.values)

    #now predict on the validation of the data 

    valid_preds=model.predict_proba(x_valid)[:,1]

    #get the roc curve 

    auc=metrics.roc_auc_score(df.valid.target_values,valid_preds)

    print(f'folds={fold},auc={auc}')

    #now run the method
    if __name__=='__main__':
      for fold_  in range(5):
        run(fold_)
    
    #now this is case where we get an idea about each fold and accuracy of score 
    

```
